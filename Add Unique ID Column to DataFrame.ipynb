```python
# Import needed libraries
import pandas as pd
```

```python
# Read CSV
url = 'https://raw.githubusercontent.com/MXMcKnight/data/master/pow_total_disease_cases.csv'
df = pd.read_csv(url)
df.head(3)
```

```python
# Create 'unique_id' column
df = df.reindex(columns = ['unique_id', *df.columns.tolist()], fill_value = 0)
df.head(3)
```

```python
# Add values to 'unique_id' column
initial_value = 1
df['unique_id'] = range(initial_value, len(df) + initial_value)
df.head(3)
```
